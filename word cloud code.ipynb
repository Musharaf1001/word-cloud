```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
with open("/content/drive/My Drive/JaneEyre.txt", "r") as file:
    text = file.read()
```

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=set(STOPWORDS),
    min_font_size=10
).generate(text)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# Example with custom parameters
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",  # Change background color
    colormap="viridis",  # Use a different color scheme
    max_words=100,  # Limit the number of words displayed
    stopwords=set(STOPWORDS).union({"additional", "stopwords"}),  # Add custom stopwords
    min_font_size=8
).generate(text)

# Display the customized word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
import numpy as np
from PIL import Image

# Load the mask image
mask = np.array(Image.open("/content/drive/My Drive/mask_image.png"))

# Create a WordCloud object with the mask
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=set(STOPWORDS),
    mask=mask,  # Apply the mask
    contour_width=1,  # Optional: add a contour around the shape
    contour_color='black'  # Optional: color of the contour
).generate(text)

# Display the shaped word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
